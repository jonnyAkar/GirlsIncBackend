```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import scipy.stats as stats
from unidecode import unidecode
import re
import math
```

```python
pd.set_option('display.max_columns',1000)
pd.set_option('display.max_rows',1000)
```

```python
fname = 'May-Oct.csv'
df = pd.read_csv(fname)
df.head()
```

```python
df.iloc[:, 4] = df.iloc[:, 4].str.extract('(\d+)')

df.iloc[:,5] = df.iloc[:,5].str.extract('(\d+)')

def remove_accents(text):
    return unidecode(text)

df['First Name'] = df['First Name'].apply(remove_accents)
df['Last Name'] = df['Last Name'].apply(remove_accents)


df.head()


df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%m/%d/%Y %H:%M:%S')

columns_to_drop = ['Is there anything you want to do more of next month in Girls Inc.?',
                   'What are some ways you can practice and improve your self awareness?',
                   'What is self awareness?', 'Is there anything you want to do less of next month in Girls Inc.?',
                   df.columns[9],  df.columns[10], df.columns[11]
                  ]

df_numbers = df.drop(columns=columns_to_drop, axis=1)

df_numbers.to_csv('May-Oct_cleaned_data.csv', index=False)
```

## Make dictionary with keys as grades

```python
#Keys are grades and values are tuples with all other columns
grades_to_values = {}

for index,row in df.iterrows():
    grade = row.iloc[4]
    
    if grade in grades_to_values:
        grades_to_values[grade].append(tuple(row.iloc[:]))
    else:
        grades_to_values[grade] = [tuple(row.iloc[:])]
        
grades_to_values['4']
```

## Grade vs Average (blank)

Let's make a function that creates an array where index represents average (blank)

```python
grades_to_blank = {}
num_students_per_grade = {}

#Input is the column we want to compare against grades
def grades_comparison(col: int):
    #iterate through each grade
    for grade in grades_to_values.keys():
        
        #iterate through each student in the grade
        grade_total_score = 0
        number_of_students = 0
        
        for student in grades_to_values[grade]:
            score = student[col] 
            grade_total_score += int(score)
            number_of_students += 1

        num_students_per_grade[grade] = number_of_students
        grades_to_blank[grade] = round(grade_total_score/number_of_students,2)

grades_comparison(5)
```

```python
for index, column_name in enumerate(df.columns):
    print(str(index) + " " + column_name)
```

```python
#Function to show the bar graph.
def plot_average_scores(grades_to_blank, column_name):
    # Create a bar graph
    plt.figure(figsize=(3, 4))
    grades_to_blank = dict(sorted(grades_to_blank.items()))
    plt.bar(grades_to_blank.keys(), grades_to_blank.values())
    plt.xlabel('Grade')
    plt.ylabel('Average Score')
    plt.title('Average Scores for ' + column_name)
    
    for grade, score in grades_to_blank.items():
        num_students = num_students_per_grade.get(grade, 0)  # Get the number of students for the grade
        plt.text(grade, score, f"{num_students}", ha='center', va='bottom')

    plt.tight_layout()
    plt.tight_layout()

    # Show the graph
    plt.show()

#COLUMN TO COMPARE
col = 7

#call grades_comparison to make the dictionary
grades_comparison(col)
#Now plot
plot_average_scores(grades_to_blank, df.columns[col])
```
